Este es para separar los nuevos datasets con artista y nombre de canción para usarlos en la búsqueda de ids

```python
import pandas as pd

# Cargar los archivos proporcionados nuevamente
genius_file = r"C:\Users\solan\Downloads\get_data_from_songs\data\processed\split_files\part1_5000\genius_songs_part_1.csv"
spoty_file = r"C:\Users\solan\Downloads\get_data_from_songs\data\raw\1_spoty_last.csv"

genius_df = pd.read_csv(genius_file)
spoty_df = pd.read_csv(spoty_file)

# Mostrar las primeras filas y columnas para entender los datos
(genius_df.head(), genius_df.columns), (spoty_df.head(), spoty_df.columns)
```

```python
# Normalizar las columnas para evitar problemas de formato
genius_df['artist_name'] = genius_df['artist_name'].str.strip().str.lower()
genius_df['song_name'] = genius_df['song_name'].str.strip().str.lower()
spoty_df['artist_name'] = spoty_df['artist_name'].str.strip().str.lower()
spoty_df['song_name'] = spoty_df['song_name'].str.strip().str.lower()

# Reintentar la unión manteniendo solo las coincidencias exactas
merged_df = pd.merge(
    spoty_df, genius_df, 
    on=['artist_name', 'song_name'], 
    how='inner'
)

# Guardar el resultado en un archivo final
final_output_path = r"C:\Users\solan\Downloads\get_data_from_songs\data\raw\limpios_bea.csv"
merged_df.to_csv(final_output_path, index=False)

# Mostrar un resumen de los resultados
merged_df.head(), merged_df.shape


```
