# 📌 MoodTune - Backend

## 📝 Descripción
El backend de MoodTune está desarrollado con **Python (Flask)** y se encarga de manejar las solicitudes del frontend, procesar datos y realizar peticiones a la API de Spotify, además de manejar dos modelos diferentes de Machine Learning.

## 📁 Estructura del Repositorio

```text
backend/  
│── flask_session/
│── src/  
│   ├── data/            # Dataset 
│   ├── models/          # Modelos
│   ├── functions/       # Funciones auxiliares  
│   ├── routes/          # Definición de endpoints  
│── run.py               # Punto de entrada del backend  
│── requirements.txt     # Dependencias del proyecto  
│── venv/                # Entorno virtual (ignorado en Git)  


## 🚀 Tecnologías Utilizadas

- **PostgreSQL** - Base de datos para almacenar canciones y moods.
- **SQLAlchemy** - ORM para manejar las consultas.
- **FuzzyWuzzy** - Para búsqueda aproximada de nombres de canciones.
- **Render** - (Inicialmente usado para despliegue, pero ahora se ejecuta en local debido a limitaciones de RAM).
- **Flask** (`Flask==3.1.0`) - Framework para la API REST.
- **Flask-Cors** (`Flask-Cors==5.0.0`) - Manejo de **CORS** para permitir solicitudes entre frontend y backend.
- **Flask-Compress** (`Flask-Compress==1.13`) - Compresión de respuestas para optimizar el rendimiento.
- **Flask-SQLAlchemy** (`Flask-SQLAlchemy==3.1.1`) - ORM para manejar la base de datos PostgreSQL.
- **flask_session** - Manejo de sesiones en Flask.
- **requests** (`requests==2.32.3`) - Para realizar solicitudes HTTP a APIs externas.
- **python-dotenv** (`python-dotenv==1.0.1`) - Carga de variables de entorno desde archivos `.env`.
- **psycopg2-binary** (`psycopg2-binary==2.9.10`) - Conector de PostgreSQL para Python.
- **gunicorn** (`gunicorn==23.0.0`) - Servidor WSGI para ejecutar la aplicación en producción.
- **pandas** - Manipulación y análisis de datos.
- **numpy** - Operaciones numéricas eficientes.
- **torch** - Librería de Machine Learning para cálculos en GPU y modelos de IA.
- **sentence-transformers** - Modelos de embeddings para transformar texto en representaciones numéricas.
- **scikit-learn** - Algoritmos de Machine Learning y herramientas estadísticas.
- **deep-translator** - Traducción automática de texto.
- **langdetect** - Detección de idioma en textos.
- **faiss-cpu** - Biblioteca para búsqueda eficiente de vectores de alta dimensión.

## 🌍 Endpoints

### 📌 **1. Redirigir al usuario a la autenticación de Spotify**
- **Ruta:** `GET /login`
- **Descripción:** Genera la URL de autenticación de Spotify y redirige al usuario para iniciar sesión.

---

### 📌 **2. Callback de autorización de Spotify**
- **Ruta:** `GET /callback`
- **Descripción:** Spotify redirige a esta ruta después de la autenticación con un código de autorización.  

---

### 📌 **3. Refrescar el token de acceso**
- **Ruta:** `POST /refresh`
- **Descripción:** Utiliza el `refresh_token` para obtener un nuevo `access_token` sin necesidad de volver a autenticarse.

---

### 📌 **4. Filtrar Top Tracks del usuario**
- **Ruta:** `POST /filter/filtered-top-tracks`
- **Descripción:** Filtra las canciones del top de Spotify del usuario, devolviendo solo aquellas que están en el dataset.

---

### 📌 **5. Filtrar Canciones Favoritas del Usuario**

- **Ruta:** `POST /filter/filtered-favourite-tracks`
- **Descripción:** Filtra las canciones favoritas del usuario en Spotify y devuelve solo las que están en el dataset.

---

### 📌 **6. Filtrar Artistas Seguidos**
- **Ruta:** `POST /filter/filtered-followed-artists`
- **Descripción:** Filtra los artistas seguidos por el usuario en Spotify, devolviendo solo aquellos que están en el dataset.

---

### 📌 **7. Obtener Canciones Filtradas desde la Sesión**
- **Ruta:** `GET /filter/get-filtered-tracks`
- **Descripción:** Recupera las canciones filtradas que se han almacenado en la sesión del usuario.

---

### 📌 **8. Obtener Canciones por Estado de Ánimo**
- **Ruta:** `POST /songs/mood`
- **Descripción:** Busca canciones basadas en el estado de ánimo del usuario.

---

### 📌 **8. Almacenar Tracks en Sesión**
- **Ruta:** `POST /user/store-tracks`
- **Descripción:** Almacena en la sesión del usuario una lista de tracks.

---

### 📌 **9. Obtener Preferencias del Usuario**
- **Ruta:** `GET /user/preferences`
- **Descripción:** Calcula las preferencias del usuario basadas en las canciones almacenadas en la sesión.


## 🔥 Decisiones Claves

- **Optimización de Búsqueda:** Para encontrar canciones de Spotify dentro del dataset, se optó por **comparaciones aproximadas (Fuzzy Matching)** en vez de búsquedas exactas.
- **Limitaciones de RAM:** Debido a problemas con Render, el backend se ejecuta **en local** para evitar restricciones de memoria.
- **Interfaz con el Frontend:** Se diseñó la API para que el frontend (**React + TypeScript**) pueda consumir los datos fácilmente.


## 📌 Próximos Pasos

- ✅ **Mejorar la eficiencia** almacenando datos y realizando consultas en PostgreSQL.  
- ✅ **Desplegar la aplicación** en un servidor con más capacidad.  
- ✅ **Implementar retroalimentación** a través de la obtención de datos de los usuarios.  


## 🛠 Instalación y Uso
   - Clonar o hacer fork del [Repositorio de Backend](https://github.com/JCMiriam/mood_tune_back) del proyecto en tu entorno local
   
   - Crear y activar entorno virtual para el proyceto:
      Para Windows
      ```bash
      python -m venv venv
      venv\Scripts\activate
      ```

      Para macOS/Linux
      ```bash
      python3 -m venv venv
      source venv/bin/activate
      ```

   - Instalar dependencias
      ```bash
      pip install -r requirements.txt
      ```

   - Añadir variables de entorno en el .env utilizando el .env.example como referencia. Dado que estas variables serán secrets, es necesario solicitarlas a uno de los creadores del proyecto o configurarlas en una aplicación de Spotify propias

   - Incluir los modelos descargables de [este directorio](https://drive.google.com/drive/folders/1i-Mq9OqLtSa0eTa0wxmNM41MtwTsX2eK?usp=sharing) en src/models/

   - Incluir el [dataset descargable](https://drive.google.com/file/d/1oTyziRBXkNGAjplpkHzYWFI6YLh5YBKL/view?usp=sharing) en src/data/ y actualizar la ruta en DATASET_PATH del archivo dataset_loader según las necesidades del entorno

   - Ejecutar la aplicación
      ```bash
            python run.py
      ```

   - Si es necesario, actualizar o instalar dependencias faltantes

   - El servidor estará disponible en http://127.0.0.1:5000
