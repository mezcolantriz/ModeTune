# 🎵 **Lógica de Negocio y Usabilidad en MoodTune** 🎶

## 📝 **1. Introducción**
MoodTune es una aplicación de recomendación de playlists en **Spotify** que **adapta la música al estado de ánimo y los gustos del usuario**.  
Cuenta con **dos sistemas de recomendación** y un flujo de usuario optimizado para facilitar la creación de playlists personalizadas.  

---

## 🔍 **2. Sistemas de Recomendación**
MoodTune ofrece **dos formas de descubrir música**, cada una con su propia lógica:

### 🎧 **1. Recomendación basada en preferencias personales**
💡 **¿Cómo funciona?**  
- **El usuario inicia sesión en Spotify** para acceder a su historial de reproducción y canciones favoritas.  
- Se **extraen datos de sus gustos** musicales mediante la API de Spotify.  
- Se filtran canciones del dataset de MoodTune según:
  - Parámetros como la bailabilidad o el tono o timbre de la canción 🎼
  - Moods de los temas (agresivo, triste, relajado, etc) 🎶 

📌 **Resultado:** Se recomienda una lista de canciones alineada con los hábitos de escucha del usuario.

---

### ✍️ **2. Recomendación basada en análisis de moods del usuario**
💡 **¿Cómo funciona?**  
- El usuario **escribe una frase** describiendo cómo se siente. 💭  
- El modelo analiza la emoción expresada en el texto. 🧠  
- Se asigna un **estado de ánimo** a la frase. 😃😢🔥  
- Se filtran canciones del dataset de MoodTune según:
  - Coincidencia con el estado de ánimo detectado en base a los embeddings 🎭  
  - Género musical preferido 🎼  

📌 **Resultado:** Se genera una playlist acorde a la emoción del usuario.  

---

## 🎛️ **3. Filtros Aplicados en Ambos Sistemas**
Antes de mostrar las recomendaciones finales, MoodTune **aplica filtros inteligentes** para mejorar la calidad de las sugerencias dependiendo del modelo:  

✅ **Filtrado por género musical** 🎸🎹🎻  
✅ **Eliminación de canciones repetidas o no disponibles en Spotify** 🚫  
✅ **Priorización de canciones similares a las ya escuchadas (para el modelo basado en preferencias)** 🔄  
✅ **Evitar recomendaciones que no coincidan con el estado de ánimo del usuario** ⚖️  

---

## 🎯 **4. Flujo del Usuario**
Tras elegir un método de recomendación, el usuario **recibe una lista con 10 canciones recomendadas**. 🎵✨  
Desde esta vista puede:

✅ **Seleccionar las canciones que desee** (puede elegir ninguna, todas o solo algunas).  
✅ **Añadirlas a una playlist de Spotify** ya existente. 📂  
✅ **Crear una nueva playlist** en su cuenta de Spotify. 🎶🆕  

💾 **Cualquier playlist creada será completamente funcional** dentro de Spotify y podrá ser reproducida en cualquier dispositivo.  

---

## 🎨 **5. Diseño y Usabilidad**
MoodTune está diseñado para ser **intuitivo y visualmente atractivo**, manteniendo una estética cercana a **Spotify**:

🖤 **Modo oscuro con colores de acento vibrantes.**  
🎛️ **Interfaz minimalista con navegación sencilla.**  
📊 **Historial de estados de ánimo con emojis para una experiencia interactiva.**  
🎵 **Vista de previsualización de canciones para decidir antes de añadir.**  

---

## 🛠️ **6. Decisiones Técnicas Clave**
⚡ **Optimización de consultas** para evitar sobrecarga en memoria.  
🔎 **Comparación aproximada de canciones** por encima de un 85% para mejorar la precisión con la API de Spotify.  
💡 **Procesamiento en segundo plano** para agilizar la generación de playlists.
